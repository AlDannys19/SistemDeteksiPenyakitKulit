# Noise Vignetting, Problem object And Grayscalling

```python
import os
import cv2
import numpy as np
import shutil

def detect_black_background(image, kernel_size=(7, 7)):  
    _, black_background = cv2.threshold(image, 20, 5, cv2.THRESH_BINARY_INV)
    return black_background

def detect_vignetting(image):  
    sigma = 130
    rows, cols = image.shape
    X_resultant_kernel = cv2.getGaussianKernel(cols, sigma)
    Y_resultant_kernel = cv2.getGaussianKernel(rows, sigma)
    resultant_kernel = Y_resultant_kernel * X_resultant_kernel.T
    mask = 255 * resultant_kernel / np.max(resultant_kernel)
    return mask

def detect_shadows(image):
    edges = cv2.Canny(image, 300, 150)
    _, shadow = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV)
    shadow_edges = cv2.bitwise_and(shadow, shadow, mask=edges)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    shadow_edges = cv2.morphologyEx(shadow_edges, cv2.MORPH_CLOSE, kernel)
    return shadow_edges

def remove_shadows(image, mask):
    result = cv2.inpaint(image, mask, 21, cv2.INPAINT_TELEA)
    return result

def remove_black_background(image, mask):
    result = cv2.inpaint(image, mask, 7, cv2.INPAINT_TELEA)
    result[mask == 255] = 200  
    return result

def remove_vignetting(image, mask):
    normalized_mask = mask / 250
    result = np.copy(image)
    result = result / (normalized_mask + 0.01)
    result = np.clip(result, 0, 200).astype(np.uint8)
    return result

def process_images_in_folder(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.endswith(('.jpg', '.jpeg', '.png')):  
                image_path = os.path.join(root, file)
                relative_path = os.path.relpath(image_path, input_folder)
                output_file_path = os.path.join(output_folder, relative_path)
                output_file_folder = os.path.dirname(output_file_path)
                
                if not os.path.exists(output_file_folder):
                    os.makedirs(output_file_folder)
                
                image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                if image is not None:
                    shadow_mask = detect_shadows(image)
                    no_shadow = remove_shadows(image, shadow_mask)
                    
                    black_background = detect_black_background(no_shadow)
                    no_black_background = remove_black_background(no_shadow, black_background)
                    
                    vignette_mask = detect_vignetting(black_background) 
                    
                    no_vignette = remove_vignetting(no_black_background, vignette_mask)

                    cv2.imwrite(output_file_path, no_vignette)
                    print(f"Processed {image_path} and saved to {output_file_path}")

input_folder = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\data\vasc"
output_folder = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\noise\vasc"

process_images_in_folder(input_folder, output_folder)
```

# Thresholding Otsu

```python
import os
import cv2

def apply_otsu_threshold(input_folder, output_folder):
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.endswith(('.jpg', '.jpeg', '.png')):  
                input_file_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_file_path, input_folder)
                output_file_path = os.path.join(output_folder, relative_path)
                output_file_folder = os.path.dirname(output_file_path)
                
                if not os.path.exists(output_file_folder):
                    os.makedirs(output_file_folder)
                
                image = cv2.imread(input_file_path, cv2.IMREAD_GRAYSCALE)
                
                _, otsu_thresholded = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
                
                otsu_thresholded_inverted = cv2.bitwise_not(otsu_thresholded)
                
                cv2.imwrite(output_file_path, otsu_thresholded_inverted)
                print(f"Memproses {input_file_path} dan menyimpan ke {output_file_path}")

input_folder = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\noise\vasc"
output_folder = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\thres\vasc"

apply_otsu_threshold(input_folder, output_folder)
```

```python
import cv2
import os

def restore_objects_with_contours(original_path, thresholded_path, output_path):
    for filename in os.listdir(thresholded_path):
        if filename.endswith(".jpg") or filename.endswith(".png"): 
            original_image = cv2.imread(os.path.join(original_path, filename))
            thresholded_image = cv2.imread(os.path.join(thresholded_path, filename), cv2.IMREAD_GRAYSCALE)

            restored_image = cv2.bitwise_and(original_image, original_image, mask=thresholded_image)

            contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            restored_image_with_contours = restored_image.copy()
            cv2.drawContours(restored_image_with_contours, contours, -1, (0, 255, 0), 2)

            cv2.imwrite(os.path.join(output_path, filename), restored_image_with_contours)

original_folder_path = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\data\vasc"
thresholded_folder_path = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\thres\vasc"
output_folder_path = r"C:\Users\Danny\Skripsi\Dataset\data 2\validasi\restore\vasc"

if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

restore_objects_with_contours(original_folder_path, thresholded_folder_path, output_folder_path)
```

# Ektraction Fiture

```python
import cv2
import numpy as np
import os
import pandas as pd

def extract_contour(color_image):
    hsv_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2HSV)
    lower_green = np.array([35, 100, 100])
    upper_green = np.array([85, 255, 255])
    mask = cv2.inRange(hsv_image, lower_green, upper_green)
    return mask

def calculate_asymmetry(binary_image):
    binary_image = binary_image.astype(bool)
    horizontal_flip = np.fliplr(binary_image)
    diff_horizontal = binary_image & ~horizontal_flip
    vertical_flip = np.flipud(binary_image)
    diff_vertical = binary_image & ~vertical_flip
    diff_horizontal_area = np.count_nonzero(diff_horizontal)
    diff_vertical_area = np.count_nonzero(diff_vertical)
    area_total = np.count_nonzero(binary_image)

    if area_total == 0:
        return 0

    asymm_idx = ((diff_horizontal_area / area_total) + (diff_vertical_area / area_total)) / 2
    return asymm_idx

def calculate_border_irregularity(color_image):
    mask = extract_contour(color_image)
    blurred = cv2.GaussianBlur(mask, (5, 5), 0)
    contours, _ = cv2.findContours(blurred, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if len(contours) == 0:
        return None
    
    largest_contour = max(contours, key=cv2.contourArea)
    perimeter = cv2.arcLength(largest_contour, True)
    area = cv2.contourArea(largest_contour)
    
    if area == 0:
        return None
    
    border_irregularity = perimeter ** 2 / (4 * np.pi * area)
    return border_irregularity

def calculate_color_variegation(image):
    skins_r = image[:, :, 2]
    skins_g = image[:, :, 1]
    skins_b = image[:, :, 0]

    mean_r = np.mean(skins_r)
    std_r = np.std(skins_r)
    mean_g = np.mean(skins_g)
    std_g = np.std(skins_g)
    mean_b = np.mean(skins_b)
    std_b = np.std(skins_b)

    C_r = std_r / np.max(skins_r) if np.max(skins_r) != 0 else 0
    C_g = std_g / np.max(skins_g) if np.max(skins_g) != 0 else 0
    C_b = std_b / np.max(skins_b) if np.max(skins_b) != 0 else 0

    color_variegation = (C_r + C_g + C_b) / 3

    return color_variegation, mean_r, std_r, mean_g, std_g, mean_b, std_b

def calculate_diameter(image, dpi=1200):
    mask = extract_contour(image)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if len(contours) == 0:
        return 0
    
    max_contour = max(contours, key=cv2.contourArea)
    _, radius = cv2.minEnclosingCircle(max_contour)
    diameter_pixels = 2 * radius
    diameter_physical = diameter_pixels * (25.4 / dpi)
    
    return diameter_physical



def process_images(base_dir, output_csv_path):
    results = []
    class_mapping = {}  # Untuk memetakan nama kelas ke angka
    class_index = 0
    
    #Iterasi melalui semua subfolder dan file gambar
    for root, dirs, files in os.walk(base_dir):
        for class_name in dirs:
            if class_name not in class_mapping:
                class_mapping[class_name] = class_index
                class_index += 1
    
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.endswith(('.png', '.jpg', '.jpeg')):
                file_path = os.path.join(root, file)
                image = cv2.imread(file_path)
                class_name = os.path.basename(os.path.dirname(file_path))
                class_label = class_mapping[class_name]
                
                border_irregularity = calculate_border_irregularity(image)
                color_variegation, mean_r, std_r, mean_g, std_g, mean_b, std_b = calculate_color_variegation(image)
                diameter = calculate_diameter(image)
                binary_image = extract_contour(image)
                asymmetry = calculate_asymmetry(binary_image)

                # Periksa nilai-nilai yang dapat menjadi NaN dan berikan nilai default jika diperlukan
                if border_irregularity is None or np.isnan(border_irregularity):
                    border_irregularity = 0
                if np.isnan(color_variegation):
                    color_variegation = 0
                if np.isnan(mean_r):
                    mean_r = 0
                if np.isnan(std_r):
                    std_r = 0
                if np.isnan(mean_g):
                    mean_g = 0
                if np.isnan(std_g):
                    std_g = 0
                if np.isnan(mean_b):
                    mean_b = 0
                if np.isnan(std_b):
                    std_b = 0

                results.append({
                    'Class': class_name,
                    'Asymmetry': asymmetry,
                    'Border Irregularity': border_irregularity,
                    'Color Variegation': color_variegation,
                    'Diameter': diameter,
                    'Mean R': mean_r,
                    'Std R': std_r,
                    'Mean G': mean_g,
                    'Std G': std_g,
                    'Mean B': mean_b,
                    'Std B': std_b
                })
    
    df = pd.DataFrame(results)
     # Memeriksa nilai NaN sebelum menyimpan ke CSV
    print("Number of NaN values in each column:")
    print(df.isnull().sum())
    
    df.to_csv(output_csv_path, index=False)
    
    return output_csv_path

base_dir = r"C:\Users\Danny\Skripsi\Dataset\data 2\restored_images_with_contours"
output_csv_path = r"C:\Users\Danny\Skripsi\Dataset\data 2\file csv\skripsi.csv"
csv_path = process_images(base_dir, output_csv_path)
print(f'Results saved to {csv_path}')
```

# SVM non Library

```python
import numpy as np
import time

class SVM:
    def __init__(self, C=1.0, gamma=1.0, degree=3, kernel='linear'):
        self.C = C
        self.gamma = gamma
        self.degree = degree
        self.kernel = kernel
        self.alpha = None
        self.bias = None
        self.support_vectors = None
        self.support_vector_labels = None
        self.classes = None
        self.n_classes = 0

    def _linear_kernel(self, x, y):
        return np.dot(x, y)

    def _polynomial_kernel(self, x, y):
        return (1 + np.dot(x, y)) ** self.degree

    def _gaussian_kernel(self, x, y):
        return np.exp(-self.gamma * np.linalg.norm(x - y) ** 2)

    def _kernel_function(self, x, y):
        if self.kernel == 'linear':
            return self._linear_kernel(x, y)
        elif self.kernel == 'polynomial':
            return self._polynomial_kernel(x, y)
        elif self.kernel == 'gaussian':
            return self._gaussian_kernel(x, y)
        else:
            raise ValueError(f"Unsupported kernel type: {self.kernel}")

    def _decision_function(self, X):
        n_samples = X.shape[0]
        decision_values = np.zeros((n_samples, self.n_classes))

        for idx in range(self.n_classes):
            bias = self.bias[idx]
            for i in range(n_samples):
                s = bias
                for alpha, sv_y, sv in zip(self.alpha[idx], self.support_vector_labels[idx], self.support_vectors[idx]):
                    s += alpha * sv_y * self._kernel_function(sv, X[i])
                decision_values[i, idx] = s

        return decision_values

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.n_classes = len(self.classes)
        self.alpha = np.zeros((self.n_classes, X.shape[0]))
        self.bias = np.zeros(self.n_classes)
        self.support_vectors = [[] for _ in range(self.n_classes)]
        self.support_vector_labels = [[] for _ in range(self.n_classes)]

        tol = 1e-2
        passes = 0
        max_passes = 5

        print(f"Starting training for {self.n_classes} classes with {self.kernel} kernel...")
        start_time = time.time()

        for idx, cls in enumerate(self.classes):
            binary_y = np.where(y == cls, 1, -1)
            self._train_one_class(X, binary_y, idx, tol, max_passes)

        end_time = time.time()
        print(f"Training completed in {end_time - start_time:.2f} seconds")

    def _train_one_class(self, X, binary_y, idx, tol, max_passes):
        self.support_vectors[idx] = X
        self.support_vector_labels[idx] = binary_y

        m = X.shape[0]
        passes = 0

        while passes < max_passes:
            num_changed_alphas = 0
            for i in range(m):
                E_i = self._decision_function(X)[i, idx] - binary_y[i]
                if (binary_y[i] * E_i < -tol and self.alpha[idx, i] < self.C) or (binary_y[i] * E_i > tol and self.alpha[idx, i] > 0):
                    j = np.random.choice([x for x in range(m) if x != i])
                    E_j = self._decision_function(X)[j, idx] - binary_y[j]
                    alpha_i_old = self.alpha[idx, i]
                    alpha_j_old = self.alpha[idx, j]

                    if binary_y[i] != binary_y[j]:
                        L = max(0, self.alpha[idx, j] - self.alpha[idx, i])
                        H = min(self.C, self.C + self.alpha[idx, j] - self.alpha[idx, i])
                    else:
                        L = max(0, self.alpha[idx, i] + self.alpha[idx, j] - self.C)
                        H = min(self.C, self.alpha[idx, i] + self.alpha[idx, j])

                    if L == H:
                        continue

                    eta = 2 * self._kernel_function(X[i], X[j]) - self._kernel_function(X[i], X[i]) - self._kernel_function(X[j], X[j])
                    if eta >= 0:
                        continue

                    self.alpha[idx, j] -= binary_y[j] * (E_i - E_j) / eta
                    self.alpha[idx, j] = np.clip(self.alpha[idx, j], L, H)

                    if abs(self.alpha[idx, j] - alpha_j_old) < tol:
                        self.alpha[idx, j] = alpha_j_old
                        continue

                    self.alpha[idx, i] += binary_y[i] * binary_y[j] * (alpha_j_old - self.alpha[idx, j])

                    b1 = self.bias[idx] - E_i - binary_y[i] * (self.alpha[idx, i] - alpha_i_old) * self._kernel_function(X[i], X[i]) - binary_y[j] * (self.alpha[idx, j] - alpha_j_old) * self._kernel_function(X[i], X[j])
                    b2 = self.bias[idx] - E_j - binary_y[i] * (self.alpha[idx, i] - alpha_i_old) * self._kernel_function(X[i], X[j]) - binary_y[j] * (self.alpha[idx, j] - alpha_j_old) * self._kernel_function(X[j], X[j])

                    if 0 < self.alpha[idx, i] < self.C:
                        self.bias[idx] = b1
                    elif 0 < self.alpha[idx, j] < self.C:
                        self.bias[idx] = b2
                    else:
                        self.bias[idx] = (b1 + b2) / 2

                    num_changed_alphas += 1

            print(f"Pass {passes+1}/{max_passes}, num_changed_alphas: {num_changed_alphas}")

            if num_changed_alphas == 0:
                passes += 1
            else:
                passes = 0

    def predict(self, X):
        decision_values = self._decision_function(X)
        predictions = np.argmax(decision_values, axis=1)
        return predictions
```
